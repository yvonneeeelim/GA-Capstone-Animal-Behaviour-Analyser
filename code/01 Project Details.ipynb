### Problem Statement

“How can we help new dog owners understand their dog’s overall emotions to foster healthy development of dogs?"

Key Emotions of A Dog
- Happy: They feel happy when there are social interactions with other canines, and able to play with their pet owners.
- Relaxed: They feel relaxed when they are in a safe and comfortable environment.
- Angry: They feel angry when they felt threatened and when their territory is occupied.
- Sad: They feel sad when there is a change in environment or when the pet owner is making them confused with their actions.

Dogs communicate their emotions through their body language
- Facial expressions: Especially through their eyes and mouth.
- Body Posture
- Ears
- Tail
